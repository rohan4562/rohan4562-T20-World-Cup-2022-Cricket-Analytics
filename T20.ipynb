# âš”ï¸ The Cricket Strategy 

Tony Sharma builds the team based on **specific player roles and performance metrics**:

### 1. Openers / Power Hitters
- Role: Start aggressively, utilize powerplay
- Metrics:
  - Batting Average
  - Strike Rate
  - Boundary %
  - At least 50+ runs partnership in first 5 overs
- Example: Jos Buttler, Rilee Rossouw


### 2. Middle Order / Anchors
- Role: Stabilize innings and accelerate when needed
- Metrics:
  - High Batting Average
  - Balls Faced
  - Ability to shift gears
- Example: Virat Kohli, Suryakumar Yadav

### 3. Finishers
- Role: Rescue in collapse OR hit at death overs
- Metrics:
  - High Strike Rate
  - Batting Average > 20
- Type: Preferably batting all-rounder

### 4. All-Rounders (Lower Middle Order)
- Role: Contribute with bat and ball at death
- Metrics:
  - Batting Avg > 15, Strike Rate > 140
  - Bowling Economy < 7
  - Bowling Strike Rate < 20


### 5. Fast Bowlers / Specialists
- Role: Take wickets and bowl dot balls
- Metrics:
  - Bowling Strike Rate < 16
  - Dot Ball % > 40%
- Example: Express pace bowlers to shake opposition


```python
# import necessary libraries

import pandas as pd
import json
```

##  Process Match Results

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
# data
```

We extract the `matchSummary` key from the loaded JSON and convert it into a DataFrame for easy handling.
     
     : Converts nested data into structured table format.

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

### Rename 'scorecard' column to 'match_id'  
To maintain consistency and enable joins across datasets, we rename the 'scorecard' column to 'match_id'.


: Ensures a unique primary key for each match.


```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

**Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables**

We map each match (Team1 vs Team2) to its `match_id`. This is necessary to link batting records with match summary.

```python
# {
#     "Namibia Vs Sri Lanka" : "T20I # 1823",
#     "Sri Lanka Vs Namibia" : "T20I # 1823",
#     "Netherlands Vs	U.A.E." : "T20I # 1825",
#     "U.A.E. Vs Netherlands" : "T20I # 1825"
#     ...
    
# }
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
#match_ids_dict
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

### Load Batting Summary JSON  
This file contains all player-level batting records for every match. We begin by reading it.

         : Loads match-wise batting records for analysis.


```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data = json.load(f)
```

```python
#data
```

We iterate over every match and merge all batting records into one flat list.

        : Converts nested JSON structure to flat DataFrame.

```python
all_records = []
for match in data:
    all_records.extend(match['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

### Add Out/Not-Out Column  
Based on the `dismissal` column, we add a new column indicating whether a batsman was out or not.

             : Enables analysis of batting performance and dismissal status.

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
```

```python
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

### Add Match ID to Batting Data  
We use the `match_ids_dict` to add a `match_id` column to the batting records.

        : Enables linking batting stats with match details.

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('Ã¢â‚¬', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
df_batting.shape
```

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```

###  Load Bowling Summary JSON

        : Brings raw bowling records into memory for analysis.


```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```
